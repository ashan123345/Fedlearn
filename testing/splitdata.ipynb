```python
import pandas as pd
import numpy as np
```

```python

# Load the dataset
data=pd.read_csv('farmighamdataset.csv')


# Define age groups
def age_group(age):
    if 20 <= age <= 39:
        return 'Young Adults (20-39)'
    elif 40 <= age <= 59:
        return 'Middle-Aged Adults (40-59)'
    elif age >= 60:
        return 'Older Adults (60+)'
    else:
        return 'Unknown'
```

```python
# Add a new column for age group classification
data['AgeGroup'] = data['age'].apply(age_group)

# Filter out 'Unknown' age groups (if any)
data = data[data['AgeGroup'] != 'Unknown']

# Split the dataset into three separate datasets
client_datasets = {
    'young_adults.csv': data[data['AgeGroup'] == 'Young Adults (20-39)'].drop(columns=['AgeGroup']),
    'middle_aged_adults.csv': data[data['AgeGroup'] == 'Middle-Aged Adults (40-59)'].drop(columns=['AgeGroup']),
    'older_adults.csv': data[data['AgeGroup'] == 'Older Adults (60+)'].drop(columns=['AgeGroup'])
}
```

```python
# Save the datasets to separate CSV files
for filename, dataset in client_datasets.items():
    dataset.to_csv(filename, index=False)
    print(f"Saved {filename} with {len(dataset)} records.")

# Summary
print("\nDataset distribution:")
for group, dataset in client_datasets.items():
    print(f"{group}: {len(dataset)} individuals")
```
